```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import time
```

```python
# URL of page to be scraped
url = 'http://www.centuryclinical.com/sponsors.html'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('ul', class_="content_list_sponsors")
print(results)
```

```python
sponsors=[]
print(len(results))
for i in range(1, len(results)):
#     print(results[i])
    for ul in results[i]:
#         print(ul)
        if (re.search("^<li>.*</li>$", str(ul) )):
            sponsor = re.sub('</li>','',re.sub('<li>','',str(re.search("^<li>.*</li>$", str(ul) ))))
            sponsors.append(sponsor[39:-2])            
# print(sponsors)
sponsors_dict = {'Sponsors':sponsors}
print(sponsors_dict)
df_sponsors = pd.DataFrame(data=sponsors_dict)
df_sponsors.head()
```

```python
df_sponsors.sort_values('Sponsors')
```

```python
urls=[]
for sponsor in sponsors:
    if sponsor in ['Medivir','Mitsubishi','Novartis','Sanofi-Synthelabo','Covance','Vector','BioPharm',
                  'Watson','Tularik','Transport','Searle','Regeneron','PPD','NPS','NiCox','MedPace','ICON']:
        
        url = 'https://clinicaltrials.gov/ct2/results?spons='
        url = f'{url}{sponsor}'
        urls.append(url)
        print(url)
# 'Therimmune','Winston','Acurian','MTRA','Omnicare',
```

```python
for url in urls:
    print(f'{url} is started')
    time.sleep(2)
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'lxml')
    # tables = soup.find_all('table')
    tables = soup.find('table', attrs={'class':'stripe row-border compact'})
    table_rows = tables.find_all('tr')
#     print(len(tables))
#     print(tables)
    l = []
    dfs=[]
    print(f'{url} is completed')
    for tr in table_rows:
        try:
            
            td = tr.find_all('td')
            row = [tr.text for tr in td]
            l.append(row)
            df_data = pd.DataFrame(l)
            df_data.drop(0,axis=1,inplace=True)
            df_data.columns=['NCT Number','Status','Title','Conditions','Interventions','Study Type','Phases',
                             'Sponsor/Collaborators','Funded Bys','Study Results','Acronym',
                             'Outcome Measures','Gender','Age','Enrollment',
                             'Study Designs','Other IDs','Start Date','Primary Completion Date','Completion Date',
                             'First Posted','Results First Posted','Last Update Posted','Locations','Study Documents']
            df_data.head(2)
            dfs.append(df_data)
    
        except Exception as e:
            print(e)

```

```python
print(len(dfs))
```

```python
df_all_sponsors = pd.concat(dfs)
```
